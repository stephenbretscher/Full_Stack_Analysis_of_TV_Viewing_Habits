```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import requests
import time
import re
import time
import random
```

```python
show_title_p = []
show_year_p = []
imdb_rating_p = []
real_rating_p =[]
prime_p = []
netflix_p = []
disney_p = []
hulu_p = []
```

```python
url2 = 'https://reelgood.com/tv/source/amazon?filter-year_end=2021&filter-year_start=2016'
executable_path = {'executable_path': ChromeDriverManager().install()}
browser2 = Browser('chrome', **executable_path, headless=False)
browser2.visit(url2)
```

```python
try:
    while True:
        time.sleep(float("{0:.2f}".format(random.uniform(1, 3))))
        browser2.click_link_by_partial_text('Load More')
except:
    print("ready to scrape")
```

```python
html2 = browser2.html
soup = bs(html2, 'html.parser')

title_col = (soup.find_all('td', class_='css-1u7zfla e126mwsw1'))

e_else_col = (soup.find_all('td', class_='css-1u11l3y'))


```

```python
for i in range(len(e_else_col)):
    if i % 4 == 0:
        show_year_p.append(str(e_else_col[i]).split("</")[0].split(">")[1])
       
```

```python
show_year_p
```

```python
len(show_year_p)
```

```python
for i in range(2, len(e_else_col)):
    if (i + 2) % 4 == 0:
        try:
            imdb_rating_p.append(str(e_else_col[i]).split("</b")[0].split(">")[5])
        except:
            imdb_rating_p.append(str(e_else_col[i]).split("</b")[0].split(">")[3])
```

```python
imdb_rating_p
```

```python
len(imdb_rating_p)
```

```python
for i in range(3, len(e_else_col)):
        if (i + 1) % 4 == 0:
            try:
                real_rating_p.append(str(e_else_col[i]).split("</b")[0].split(">")[5])
            except:
                real_rating_p.append(str(e_else_col[i]).split("</b")[0].split(">")[3])
    
```

```python
real_rating_p
```

```python
len(real_rating_p)
```

```python
for i in range(len(title_col)):
    show_title_p.append((str(title_col[i])).split("</")[0].split("\">")[2])
```

```python
show_title_p
```

```python
len(show_title_p)
```

```python
for i in range(len(show_title_p)):
    prime_p.append(True)
```

```python
prime_p
```

```python
len(prime_p)
```

```python
for i in range(len(show_title_p)):
    netflix_p.append(False)
    disney_p.append(False)
    hulu_p.append(False)
    
```

```python
len(netflix_p)
```

```python
len(disney_p)
```

```python
len(hulu_p)
```

```python
prime_whole = [show_title_p, show_year_p, imdb_rating_p, real_rating_p, prime_p, netflix_p, disney_p, hulu_p]
```

```python
browser2.quit()
```

```python
df = pd.DataFrame(prime_whole)
df.to_csv('prime_whole.csv', index=False)
```
