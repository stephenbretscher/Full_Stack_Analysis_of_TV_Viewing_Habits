```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
tv_show_csv = "CSVs/tv_shows.csv"

tv_shows_df = pd.read_csv(tv_show_csv)
```

```python
tv_shows_df
```

```python
tv_shows_2019 = tv_shows_df.loc[tv_shows_df['Year'] == 2019]
tv_shows_2019
```

```python
tv_shows_2020 = tv_shows_df.loc[tv_shows_df['Year'] == 2020]
tv_shows_2020
```

```python
tv_shows_2020.sum(axis = 0)
```

```python
netflix_df = tv_shows_2020.loc[tv_shows_2020['Netflix'] == 1]
netflix_df
```

```python
hulu_df = tv_shows_2020.loc[tv_shows_2020['Hulu'] == 1]
hulu_df.head()
```

```python
prime_df = tv_shows_2020.loc[tv_shows_2020['Prime Video'] == 1]
prime_df.head()
```

```python
disney_df = tv_shows_2020.loc[tv_shows_2020['Disney+'] == 1]
disney_df
```

```python
connection_string = "postgres:postgres@localhost:5432/project-2"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
tv_shows_2020.to_sql(name='tv_shows', con=engine, if_exists='append', index=False)

```
